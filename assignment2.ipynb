```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("House Price India.csv")

data
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Price'])
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data['Location'])
plt.title('Countplot of Location')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Area', y='Price', data=data)
plt.title('Scatter Plot between Area and Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()

```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
descriptive_stats = data.describe()

data.dropna(inplace=True)

median_bedrooms = data['Bedrooms'].median()
data['Bedrooms'].fillna(median_bedrooms, inplace=True)

data.to_csv("Cleaned_House_Price_India.csv", index=False)
```
